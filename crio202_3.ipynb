##Portfolio Construction


```python
monthly_returns.columns
```

```python
selected_funds = [
    'TCS.NS',
    'HDFCBANK.NS',
    'RELIANCE.NS',
    'HINDUNILVR.NS'
]
```

```python
weights = np.array([0.25, 0.25, 0.30, 0.20])
```

```python
portfolio_returns = monthly_returns[selected_funds].dot(weights)
portfolio_returns.head()
```

```python
portfolio_annual_return = portfolio_returns.mean() * 12
portfolio_annual_return
```

```python
cov_matrix = monthly_returns[selected_funds].cov() * 12

portfolio_volatility = np.sqrt(
    np.dot(weights.T, np.dot(cov_matrix, weights))
)

portfolio_volatility
```

```python
portfolio_sharpe = (portfolio_annual_return - risk_free_rate) / portfolio_volatility
portfolio_sharpe
```

```python
def portfolio_performance(weights, returns, rf=0.06):
    annual_return = np.sum(returns.mean() * weights) * 12
    cov_matrix = returns.cov() * 12
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe = (annual_return - rf) / volatility

    return annual_return, volatility, sharpe
```

```python
portfolio_performance(weights, monthly_returns[selected_funds])
```

```python
portfolio_cumulative = (1 + portfolio_returns).cumprod()

plt.figure(figsize=(10,6))
portfolio_cumulative.plot()
plt.title("Portfolio Growth Over Time")
plt.ylabel("Value of ₹1 Invested")
plt.xlabel("Date")
plt.show()
```

```python
comparison = cumulative_returns[selected_funds].copy()
comparison['Portfolio'] = portfolio_cumulative

plt.figure(figsize=(10,6))
comparison.plot()
plt.title("Portfolio vs Individual Stocks")
plt.ylabel("Growth of ₹1")
plt.show()
```

##Monte Carlo Simulation

```python
num_simulations = 1000
num_years = 10

months = num_years * 12
```

```python
mean_return = portfolio_returns.mean()
std_dev = portfolio_returns.std()

mean_return, std_dev
```

```python
np.random.seed(42)

simulation_results = np.zeros((months, num_simulations))

for i in range(num_simulations):
    monthly_returns_sim = np.random.normal(
        mean_return,
        std_dev,
        months
    )
    simulation_results[:, i] = np.cumprod(1 + monthly_returns_sim)
```

```python
plt.figure(figsize=(10,6))
plt.plot(simulation_results)
plt.title("Monte Carlo Simulation of Portfolio Value")
plt.xlabel("Months")
plt.ylabel("Portfolio Value (₹1 Invested)")
plt.show()
```

```python
final_values = simulation_results[-1, :]
final_values[:10]
```

```python
best_case = np.percentile(final_values, 95)
worst_case = np.percentile(final_values, 5)
expected_case = np.percentile(final_values, 50)

best_case, expected_case, worst_case
```

##Risk insight — Probability of loss

```python
probability_of_loss = np.mean(final_values < 1)
probability_of_loss
```

```python
plt.figure(figsize=(8,6))
plt.hist(final_values, bins=50)
plt.axvline(best_case, linestyle='--')
plt.axvline(expected_case, linestyle='--')
plt.axvline(worst_case, linestyle='--')

plt.title("Distribution of Final Portfolio Values")
plt.xlabel("Portfolio Value after 10 Years")
plt.ylabel("Frequency")
plt.show()
```

##Portfolio Optimization

```python
import random
```

```python
returns_subset = monthly_returns[selected_funds]

mean_returns = returns_subset.mean() * 12
cov_matrix = returns_subset.cov() * 12
```

```python
def random_portfolios(num_portfolios, mean_returns, cov_matrix, rf=0.06):
    results = np.zeros((3, num_portfolios))
    weights_record = []

    for i in range(num_portfolios):
        weights = np.random.random(len(mean_returns))
        weights /= np.sum(weights)
        weights_record.append(weights)

        portfolio_return = np.sum(mean_returns * weights)
        portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
        sharpe_ratio = (portfolio_return - rf) / portfolio_volatility

        results[0,i] = portfolio_return
        results[1,i] = portfolio_volatility
        results[2,i] = sharpe_ratio

    return results, weights_record
```

```python
num_portfolios = 5000

results, weights_record = random_portfolios(
    num_portfolios,
    mean_returns,
    cov_matrix
)
```

```python
max_sharpe_idx = np.argmax(results[2])
min_vol_idx = np.argmin(results[1])

max_sharpe_return = results[0, max_sharpe_idx]
max_sharpe_vol = results[1, max_sharpe_idx]
max_sharpe_weights = weights_record[max_sharpe_idx]

min_vol_return = results[0, min_vol_idx]
min_vol_vol = results[1, min_vol_idx]
min_vol_weights = weights_record[min_vol_idx]
```

```python
print("📌 Maximum Sharpe Ratio Portfolio")
print("Return:", round(max_sharpe_return, 3))
print("Risk:", round(max_sharpe_vol, 3))
print("Weights:")
for fund, w in zip(selected_funds, max_sharpe_weights):
    print(f"  {fund}: {round(w,2)}")

print("\n📌 Minimum Volatility Portfolio")
print("Return:", round(min_vol_return, 3))
print("Risk:", round(min_vol_vol, 3))
print("Weights:")
for fund, w in zip(selected_funds, min_vol_weights):
    print(f"  {fund}: {round(w,2)}")
```

##Efficient Frontier

```python
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c=results[2,:])
plt.colorbar(label="Sharpe Ratio")

plt.scatter(max_sharpe_vol, max_sharpe_return, marker='*', s=200, label='Max Sharpe')
plt.scatter(min_vol_vol, min_vol_return, marker='*', s=200, label='Min Volatility')

plt.xlabel("Risk (Volatility)")
plt.ylabel("Return")
plt.title("Efficient Frontier")
plt.legend()
plt.show()
```

##Summary Report

```python
def generate_investment_report(
    investment,
    years,
    risk_profile,
    funds,
    weights,
    exp_return,
    exp_risk,
    exp_sharpe,
    future_value
):

    report = f"""
    MUTUAL FUND INVESTMENT PLAN SUMMARY
    ----------------------------------

    Investment Amount      : ₹{investment:,.0f}
    Investment Horizon     : {years} years
    Risk Profile           : {risk_profile.capitalize()}

    Recommended Portfolio Allocation:
    """

    for fund, w in zip(funds, weights):
        report += f"    - {fund}: {w*100:.1f}%\n"

    report += f"""

    Expected Portfolio Performance:
    --------------------------------
    Expected Annual Return : {exp_return*100:.2f}%
    Expected Risk (Volatility) : {exp_risk*100:.2f}%
    Sharpe Ratio : {exp_sharpe:.2f}

    Projected Investment Value:
    ---------------------------
    After {years} years, your investment of ₹{investment:,.0f}
    is expected to grow to approximately:

    👉 ₹{future_value:,.0f}

    Disclaimer:
    ----------
    This projection is based on historical data and simulations.
    Actual returns may vary due to market conditions.
    """

    return report
```

```python
final_report = generate_investment_report(
    initial_investment,
    investment_horizon_years,
    risk_profile,
    selected_funds,
    chosen_weights,
    exp_return,
    exp_risk,
    exp_sharpe,
    future_value
)

print(final_report)
```
